```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df = df.drop(['deck', 'who', 'adult_male', 'parch', 'sibsp'], axis=1)

df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
counts = df['sex'].value_counts()
print("male:", counts['male'])
print("female:", counts['female'])
sns.countplot(data=df, x='sex', hue = 'sex')
plt.show()
```

```python
class_counts = df.groupby('class')['sex'].value_counts()
print(class_counts)
sns.countplot(data=df ,x='sex',hue='class')
```

```python
male_df = df[df['sex'] == 'male']
sns.countplot(data=male_df, x='class')
male_class_counts = male_df['class'].value_counts()
print(male_class_counts)
```

```python
female_df = df[df['sex'] == 'female']
sns.countplot(data=female_df, x='class')
female_class_counts = female_df['class'].value_counts()
print(female_class_counts)
```
